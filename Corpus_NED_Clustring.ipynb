```python
df.arlabel.dropna()
```

```python
df = df[df['arlabel'].notna()]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df.arlabel)
```

3. Apply K-means clustering on the feature vectors with the objective of getting 2 clusters as similar and dissimilar


```python
true_k = 1000
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=500, n_init=1)
model.fit(X)
```

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
terms
```
